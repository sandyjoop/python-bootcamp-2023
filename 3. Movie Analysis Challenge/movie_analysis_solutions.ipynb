**Step 4: Exploring Data Using Functions (30 minutes) - 15 points**
1. Define a function to calculate the average duration of movies in a specific genre.
2. Create a unit test for the above function using pytest in another python file and test the function.
3. Use an external library (e.g., NumPy) to calculate the correlation between year and duration.

```python
def genre_avg_duration(genre):
    # Select subset of records with the specified genre
    genre_subset = movies_df[movies_df['first_genre'] == genre]
    
    # Calculate the average duration of movies in the genre
    avg_duration = genre_subset['duration'].mean()
    
    return avg_duration

avg_duration = genre_avg_duration('Comedy')
print("Average duration is: {}".format(avg_duration))
```

**Step 5: DataFrame Manipulations (30 minutes) - 15 points**
1. Sort the dataset based on year.
2. Use a loop to find the most common director in the dataset.

```python
# Sort the DataFrame based on the 'Year' column in ascending order
sorted_df = movies_df.sort_values(by='year')

# Display the sorted DataFrame
print(sorted_df)
```

```python
# Create a dictionary to count director occurrences
director_count = {}

# Loop through each row in the DataFrame
for index, row in movies_df.iterrows():
    directors = [director.strip() for director in row['director'].split(',')]  # Strip spaces and split by comma
    # Use only the first director in the list
    director = directors[0]
    
    # Update director_count dictionary
    if director in director_count:
        director_count[director] += 1
    else:
        director_count[director] = 1

# Find the most common director
most_common_director = max(director_count, key=director_count.get)

print(f"The most common director is: {most_common_director}")
```

```python
# Create a dictionary to count director occurrences
director_count = {}

# Loop through each row in the DataFrame
for index, row in movies_df.iterrows():
    directors = [director.strip() for director in row['director'].split(',')]  # Strip spaces and split by comma
    
    # Loop through all directors in the list
    for director in directors:
        # Update director_count dictionary
        if director in director_count:
            director_count[director] += 1
        else:
            director_count[director] = 1

# Find the most common director
most_common_director = max(director_count, key=director_count.get)

print(f"The most common director is: {most_common_director}")
```

**Step 6: Advanced Data Analysis (40 minutes) - 15 points**
1. Load another ratings data in "IMDB ratings" and merge it with the original dataset.
2. Calculate the average rating for each genre and visualize the results.
3. Identify the 5 top-rated movies and display their details.

```python
# Assuming you have two DataFrames: movies_df and ratings_df
ratings_df = pd.read_csv("IMDb ratings.csv")
ratings_df.head()

# Merge the two DataFrames based on the 'imdb_title_id' column
merged_df = movies_df.merge(ratings_df, on='imdb_title_id', how='inner')

# 'how' parameter specifies the type of merge you want to perform (inner, outer, left, or right)
# 'on' parameter specifies the column to use for merging

# Display the merged DataFrame
merged_df.head()
```

```python
# Step 1: Group by 'genre' and calculate the average votes
genre_avg_votes = merged_df.groupby('first_genre')['weighted_average_vote'].mean().reset_index()

# Step 2: Extract distinct genre values
distinct_genres = genre_avg_votes['first_genre']

# Step 3: Create a bar chart to visualize the results
plt.figure(figsize=(10, 6))
plt.bar(distinct_genres, genre_avg_votes['weighted_average_vote'])
plt.xlabel('Genre')
plt.ylabel('Average Votes')
plt.title('Average Votes per Genre')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Find the top 5 rated movies
top_rated_movies = merged_df.nlargest(5, 'weighted_average_vote')

# Display details of the top 5 rated movies
top_rated_movies.head()
```

**Challenging Bonus (20 minutes) - 20 points**
1. Visualize the changing popularity of genres over the years.


```python
# Determine the top 5 genres based on your criteria
genre_avg_rating = merged_df.groupby('first_genre')['weighted_average_vote'].mean().sort_values(ascending=False)
top_5_genres = genre_avg_rating.head(5).index

# Filter the data to include only the top 5 genres
filtered_df = merged_df[merged_df['first_genre'].isin(top_5_genres)]

# Group by 'year' and 'first_genre' and calculate the average rating
genre_avg_rating = filtered_df.groupby(['year', 'first_genre'])['weighted_average_vote'].mean().unstack()

# Create a line plot for the top 5 genres
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")  # Optional: Set a background grid for the plot

for genre in top_5_genres:
    plt.plot(genre_avg_rating.index, genre_avg_rating[genre], label=genre)

plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Variation of Average Ratings for Top 5 Genres Over the Years')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Group by 'year' and 'first_genre' and calculate the average weighted average vote
genre_avg_vote = merged_df.groupby(['first_genre', pd.cut(merged_df['year'], range(1920, 2031, 10))]).agg({'weighted_average_vote': 'mean'})

# Reset the index to make the bins and genres as columns
genre_avg_vote = genre_avg_vote.reset_index()

# Pivot the data to have genres as columns
genre_avg_vote_pivot = genre_avg_vote.pivot(index='year', columns='first_genre', values='weighted_average_vote')

# Select the top 5 genres
top_5_genres = genre_avg_vote['first_genre'].value_counts().head(5).index

# Filter the data to include only the top 5 genres
genre_avg_vote_pivot = genre_avg_vote_pivot[top_5_genres]

# Create a column chart for the top 5 genres
ax = genre_avg_vote_pivot.plot(kind='bar', figsize=(12, 6))
plt.xlabel('Year Bins (Every 10 Years)')
plt.ylabel('Average Weighted Average Vote')
plt.title('Changing Popularity of Top 5 Genres Over Time (in 10-Year Bins)')
plt.grid(True)
plt.legend(title='Genre')

plt.show()
```
